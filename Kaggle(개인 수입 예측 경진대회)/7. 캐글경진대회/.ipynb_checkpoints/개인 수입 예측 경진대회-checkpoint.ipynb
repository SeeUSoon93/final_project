## 결측치 제거
- null대신 ?이 들어가있음

```python
train['workclass'].value_counts()
```

```python
train['native-country'].value_counts()
```

```python
train['occupation'].value_counts()
```

```python
# ?의 비율이 적으므로 그냥 삭제해버리자
notnull_train = train[(train['workclass'] == '?')&
      (train['occupation'] == '?')&
      (train['native-country']=='?')]
```

```python
train.drop(index=notnull_train.index, axis=0, inplace=True)
```

```python
# test도 ?지우기
notnull_test = test[(test['workclass'] == '?')&
      (test['occupation'] == '?')&
      (test['native-country']=='?')]
test.drop(index=notnull_test.index, axis=0, inplace=True)
```

### value값 띄어쓰기 제거

```python
# 띄어쓰기 제거함
for i in train.columns : 
    if train[f'{i}'].dtype == 'object' :
        train[f'{i}'] = train[f'{i}'].apply(lambda x: x.replace(' ', '')) # 제거하는 함수
    else :
        train[f'{i}'] = train[f'{i}']
```
